```python
import os
import pandas as pd

# Path to the folder containing the CSV files
folder_path = '/content/drive/MyDrive/VR Mini Project'

# Get a list of all CSV files in the folder
csv_files = [f for f in os.listdir(folder_path) if f.endswith('.csv')]
```

```python
csv_files
```

```python
# Create an empty list to hold individual DataFrames
df_list = []

# Loop through and read each CSV file
for file in csv_files:
    file_path = os.path.join(folder_path, file)
    df = pd.read_csv(file_path)
    df_list.append(df)

# Combine all DataFrames into a single DataFrame
combined_df = pd.concat(df_list, ignore_index=True)

# Save the combined DataFrame to a new CSV file
combined_df.to_csv('full_data_curated.csv', index=False)

print(f"Combined {len(csv_files)} files into 'combined_output.csv'")
```
